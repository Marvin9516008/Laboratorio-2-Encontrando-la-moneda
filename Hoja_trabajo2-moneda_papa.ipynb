```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("C:/papa/IMG_9391.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
# Imagen con escala de grise
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Aplicando suavizado Gaussiano
gaussiana = cv2.GaussianBlur(gris, (5,5), 0)
# Convertiendo a escala de grises
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
 
# Aplicar suavizado Gaussiano
gauss = cv2.GaussianBlur(gris, (5,5), 0)
 
#cv2.imshow("suavizado", gauss)
plt.imshow(gauss)
# Detectamos los bordes con Canny
canny = cv2.Canny(gauss, 50, 150)
 
#cv2.imshow("canny", canny)
plt.imshow(canny)
# Buscamos los contornos
(contornos,_) = cv2.findContours(canny.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
 
# Mostramos el número de monedas por consola
print("He encontrado {} objetos".format(len(contornos)))
 
cv2.drawContours(img,contornos,-1,(0,0,255), 2)
#cv2.imshow("contornos", original)
plt.imshow(img)
cv2.waitKey(0)
alto = img.shape[0]
ancho = img.shape[1]
lienzo = np.zeros((alto, ancho, 3)) # resultado a color.
for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue  = pixel[2]
        green = pixel[1]
        red   = pixel[0]
        
        lienzo[i, j] = [blue, 0, 0]
media_ponderada = np.zeros((alto, ancho, 1)) # escala de grises
for i in range(0, alto):
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue  = pixel[2]
        green = pixel[1]
        red   = pixel[0]
        
        media_ponderada[i, j] = int(0.299* blue + 0.58 * green + 0.114 * red)
cv2.imwrite("gris_ponderado.jpg", media_ponderada)
gris = cv2.imread("gris_ponderado.jpg")
gris = cv2.cvtColor(gris, cv2.COLOR_BGR2RGB)
plt.imshow(gris)
plt.show()
```

```python
_,binary = cv2.threshold(gray,100,255,cv2.THRESH_BINARY)
plt.imshow(binary)
```

```python
contours,hierarchy = cv2.findContours(binary,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
drawing = np.zeros((gray.shape[0], gray.shape[1], 3), dtype=np.uint8)
CountersImg = cv2.drawContours(drawing,contours, -1, (0,255,0),3)
plt.imshow(CountersImg)
cv2.waitKey(0)
```
